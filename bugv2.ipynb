```python
import pandas as pd
import re
import numpy as np
from typing import Dict, List, Optional, Tuple
import warnings
warnings.filterwarnings('ignore')

class BugDescriptionParser:
    """
    Complete bug description parser for Colab that creates a structured DataFrame.
    """

    def __init__(self):
        """Initialize the parser."""
        self.initialize_patterns()

    def initialize_patterns(self):
        """Initialize all regex patterns for extraction."""
        self.template_patterns = {
            'USER_FEEDBACK_REPORT': r'^### User feedback report',
            'CONTEXT_ATN_BUG': r'^# Context',
            'TEST_CASE_BUG': r'Test Case Link:|test effort http',
            'MEDIA_MOVIE_ISSUE': r'^(The movie|Open entity page|Title of movie)',
            'CRASH_CLUSTER': r"(There's a crash cluster|crash cluster belonging)",
            'WHAT_STEPS_FORMAT': r'What steps would let us observe',
            'TLDR_FORMAT': r'^TLDR:',
            'SUMMARY_FORMAT': r'^\*?\*?Summary:?\*?\*?',
            'SIGNAL_CRASH': r'^signal \d+ \(',
        }

    def parse_single_description(self, description: str) -> Dict:
        """
        Parse a single description and return a dictionary with all fields.
        Missing fields will be None (which becomes NaN in DataFrame).
        """
        # Initialize all columns with None (will become NaN in DataFrame)
        result = {
            # Basic Information
            'template_type': None,
            'issue_type': None,
            'summary': None,

            # Steps and Results
            'steps_to_reproduce': None,
            'steps_count': None,
            'expected_result': None,
            'actual_result': None,

            # Environment
            'device': None,
            'device_model': None,
            'os_version': None,
            'android_version': None,
            'build_id': None,
            'app_version': None,
            'firmware_version': None,

            # Test Information
            'test_case_id': None,
            'test_case_name': None,
            'test_case_link': None,
            'regression_id': None,

            # Error Details
            'error_type': None,
            'error_message': None,
            'stack_trace': None,
            'signal_type': None,
            'crash_id': None,

            # URLs and References
            'bug_ids': None,
            'sherlog_links': None,
            'screenshot_links': None,
            'total_urls': None,

            # User Feedback
            'feedback_id': None,
            'user_query': None,

            # Metrics
            'frequency': None,
            'reproducibility': None,
            'vote_count': None,

            # Status
            'is_regression': None,
            'is_blocking': None,
            'severity': None,
            'priority': None
        }

        # Check if description is valid
        if pd.isna(description) or not description:
            return result

        # Detect template type
        result['template_type'] = self.detect_template(description)

        # Extract all fields
        result['issue_type'] = self.extract_issue_type(description)
        result['summary'] = self.extract_summary(description)

        # Extract steps
        steps = self.extract_steps(description)
        if steps:
            result['steps_to_reproduce'] = ' | '.join(steps)
            result['steps_count'] = len(steps)

        # Extract expected and actual
        expected, actual = self.extract_expected_actual(description)
        result['expected_result'] = expected
        result['actual_result'] = actual

        # Extract environment info
        env_info = self.extract_environment(description)
        result.update(env_info)

        # Extract test info
        test_info = self.extract_test_info(description)
        result.update(test_info)

        # Extract error details
        error_info = self.extract_error_details(description)
        result.update(error_info)

        # Extract URLs
        url_info = self.extract_urls(description)
        result.update(url_info)

        # Extract metrics
        metrics = self.extract_metrics(description)
        result.update(metrics)

        return result

    def detect_template(self, description: str) -> str:
        """Detect template type."""
        for template_name, pattern in self.template_patterns.items():
            if re.search(pattern, description, re.IGNORECASE | re.MULTILINE):
                return template_name
        return 'GENERIC'

    def extract_issue_type(self, description: str) -> str:
        """Extract issue type."""
        desc_lower = description.lower()
        if 'crash' in desc_lower or 'signal' in desc_lower:
            return 'crash'
        elif 'test' in desc_lower and 'fail' in desc_lower:
            return 'test_failure'
        elif 'user feedback' in desc_lower:
            return 'user_feedback'
        elif 'bug' in desc_lower:
            return 'bug'
        return 'issue'

    def extract_summary(self, description: str) -> Optional[str]:
        """Extract summary."""
        # Try different summary patterns
        patterns = [
            r'### Description\s*\n(.*?)(?=\n###|\Z)',
            r'TLDR:\s*(.*?)(?=\n)',
            r'\*?\*?Summary:?\*?\*?\s*(.*?)(?=\n\n|\Z)',
            r'^([^\n]{10,200})(?=\n)'  # First line if meaningful
        ]

        for pattern in patterns:
            match = re.search(pattern, description, re.IGNORECASE | re.DOTALL)
            if match:
                summary = match.group(1).strip()
                if len(summary) > 10:
                    return summary[:300]  # Limit length

        # Fallback: first 200 chars
        return description[:200].strip()

    def extract_steps(self, description: str) -> List[str]:
        """Extract steps to reproduce."""
        steps = []

        # Look for steps section
        steps_section = re.search(
            r'(?:STEPS?\s*TO\s*REPRODUCE|STR:|What steps)[:\s]*(.*?)(?=\n[A-Z]{4,}:|\n\*\*|\Z)',
            description, re.IGNORECASE | re.DOTALL
        )

        if steps_section:
            content = steps_section.group(1)
            # Extract numbered steps
            numbered = re.findall(r'\n?\s*\d+[\.)]\s*(.*?)(?=\n\s*\d+[\.]|\Z)', content, re.DOTALL)
            if numbered:
                steps = [s.strip() for s in numbered if s.strip()]

        # Try to find numbered steps anywhere
        if not steps:
            numbered = re.findall(r'(?:^|\n)\s*\d+[\.)]\s*(.*?)(?=\n\s*\d+[\.]|\n\n|\Z)',
                                description, re.MULTILINE | re.DOTALL)
            if len(numbered) >= 2:
                steps = [s.strip() for s in numbered if s.strip()][:10]

        return steps

    def extract_expected_actual(self, description: str) -> Tuple[Optional[str], Optional[str]]:
        """Extract expected and actual results."""
        expected = None
        actual = None

        # Expected patterns
        exp_match = re.search(
            r'(?:EXPECTED|Expected)(?:\s*(?:result|behavior|outcome))?[\s:]+([^\n]+(?:\n(?!\n)[^\n]+)*)',
            description, re.IGNORECASE
        )
        if exp_match:
            expected = exp_match.group(1).strip()[:300]

        # Actual patterns
        act_match = re.search(
            r'(?:ACTUAL|Actual|Observed)(?:\s*(?:result|behavior|outcome))?[\s:]+([^\n]+(?:\n(?!\n)[^\n]+)*)',
            description, re.IGNORECASE
        )
        if act_match:
            actual = act_match.group(1).strip()[:300]

        return expected, actual

    def extract_environment(self, description: str) -> Dict:
        """Extract environment information."""
        env = {}

        patterns = {
            'device': r'(?:Device|Phone Model)[\s:]+([^\n]+)',
            'device_model': r'(?:Model|Device Model)[\s:]+([^\n]+)',
            'os_version': r'(?:OS Version)[\s:]+([^\n]+)',
            'android_version': r'(?:Android Version|ANDROID OS VERSION)[\s:]+([^\n]+)',
            'build_id': r'(?:Build ID|BUILD-ID|Build Number|Build)[\s:]+([^\n]+)',
            'app_version': r'(?:App Version|APK VERSION|VERSION)[\s:]+([^\n]+)',
            'firmware_version': r'(?:Firmware|Watch Firmware)[\s:]+([^\n]+)'
        }

        for key, pattern in patterns.items():
            match = re.search(pattern, description, re.IGNORECASE)
            if match:
                env[key] = match.group(1).strip()[:100]

        return env

    def extract_test_info(self, description: str) -> Dict:
        """Extract test-related information."""
        test_info = {}

        # Test case link
        link_match = re.search(r'Test Case Link:\s*(https?://[^\s\n]+)', description, re.IGNORECASE)
        if link_match:
            test_info['test_case_link'] = link_match.group(1)

        # Test case ID/number
        id_match = re.search(r'Test case number:\s*([^\n]+)', description, re.IGNORECASE)
        if id_match:
            test_info['test_case_id'] = id_match.group(1).strip()

        # Test case name
        name_match = re.search(r'Test case.*?:\s*[\d.]+\s+([^\n]+)', description, re.IGNORECASE)
        if name_match:
            test_info['test_case_name'] = name_match.group(1).strip()

        # Regression ID
        reg_match = re.search(r'regression ID:\s*\[?([^\]\n]+)', description, re.IGNORECASE)
        if reg_match:
            test_info['regression_id'] = reg_match.group(1).strip()

        return test_info

    def extract_error_details(self, description: str) -> Dict:
        """Extract error and crash details."""
        error_info = {}

        # Signal type
        signal_match = re.search(r'signal (\d+)\s*\(([^)]+)\)', description, re.IGNORECASE)
        if signal_match:
            error_info['signal_type'] = f"signal {signal_match.group(1)} ({signal_match.group(2)})"

        # Stack trace (simplified extraction)
        stack_match = re.search(r'(?:stack trace|backtrace).*?\n(.*?)(?=\n\n|\Z)',
                               description, re.IGNORECASE | re.DOTALL)
        if stack_match:
            error_info['stack_trace'] = stack_match.group(1).strip()[:500]

        # Crash ID
        crash_match = re.search(r'(?:CrashID|crash id):\s*([^\n]+)', description, re.IGNORECASE)
        if crash_match:
            error_info['crash_id'] = crash_match.group(1).strip()

        # Error message
        error_match = re.search(r'(?:error message|error):\s*([^\n]+)', description, re.IGNORECASE)
        if error_match:
            error_info['error_message'] = error_match.group(1).strip()[:200]

        return error_info

    def extract_urls(self, description: str) -> Dict:
        """Extract URLs and references."""
        url_info = {}

        # Bug IDs
        bug_ids = re.findall(r'b/(\d+)', description)
        if bug_ids:
            url_info['bug_ids'] = ', '.join(bug_ids[:5])  # Limit to 5

        # Sherlog links
        sherlogs = re.findall(r'(?:sherlog|Sherlog)[:\s]*([^\s\n]+)', description, re.IGNORECASE)
        if sherlogs:
            url_info['sherlog_links'] = ', '.join(sherlogs[:3])

        # Screenshots
        screenshots = re.findall(r'(?:screenshot|Screenshot)[:\s]*(https?://[^\s\n]+)', description, re.IGNORECASE)
        if screenshots:
            url_info['screenshot_links'] = ', '.join(screenshots[:3])

        # Total URLs count
        all_urls = re.findall(r'https?://[^\s\n]+', description)
        url_info['total_urls'] = len(all_urls)

        # Feedback ID
        feedback_match = re.search(r'listnr\.corp\.google\.com/report/(\d+)', description)
        if feedback_match:
            url_info['feedback_id'] = feedback_match.group(1)

        return url_info

    def extract_metrics(self, description: str) -> Dict:
        """Extract metrics and impact information."""
        metrics = {}

        # Reproducibility/Frequency
        repro_match = re.search(r'(?:Repeatable|Reproducible|Frequency)[\s:]+([^\n]+)', description, re.IGNORECASE)
        if repro_match:
            metrics['reproducibility'] = repro_match.group(1).strip()[:50]

        # Vote count
        vote_match = re.search(r'(?:Vote count|Votes)[\s:]+(\d+)', description, re.IGNORECASE)
        if vote_match:
            metrics['vote_count'] = int(vote_match.group(1))

        # Regression
        if re.search(r'Regression:\s*Yes', description, re.IGNORECASE):
            metrics['is_regression'] = 'Yes'
        elif re.search(r'Regression:\s*No', description, re.IGNORECASE):
            metrics['is_regression'] = 'No'

        # Severity/Priority
        severity_match = re.search(r'(?:Severity|Priority)[\s:]+([^\n]+)', description, re.IGNORECASE)
        if severity_match:
            metrics['severity'] = severity_match.group(1).strip()[:50]

        return metrics


# Main function to process CSV and create new DataFrame
def process_bug_descriptions(csv_path: str, num_descriptions: int = 10) -> pd.DataFrame:
    """
    Process bug descriptions from CSV and create a structured DataFrame.

    Args:
        csv_path: Path to the CSV file
        num_descriptions: Number of descriptions to process (default: 10)

    Returns:
        DataFrame with all extracted information as columns
    """
    # Read the CSV file
    print(f"Reading CSV file: {csv_path}")
    df = pd.read_csv(csv_path)

    # Initialize parser
    parser = BugDescriptionParser()

    # Process first N descriptions
    print(f"\nProcessing first {num_descriptions} descriptions...")
    results = []

    for i in range(min(num_descriptions, len(df))):
        if i % 2 == 0:  # Progress indicator
            print(f"Processing description {i+1}/{num_descriptions}...")

        # Get description
        description = df.iloc[i]['Description'] if 'Description' in df.columns else df.iloc[i][0]

        # Parse description
        parsed_data = parser.parse_single_description(description)

        # Add original description (truncated) for reference
        parsed_data['original_description'] = description[:500] if pd.notna(description) else None

        results.append(parsed_data)

    # Create DataFrame from results
    result_df = pd.DataFrame(results)

    print(f"\n✅ Processing complete!")
    print(f"Created DataFrame with {len(result_df)} rows and {len(result_df.columns)} columns")

    return result_df


# Example usage for Colab
if __name__ == "__main__":

    csv_filename = '/content/description_only.csv'  # Change to your filename
    result_df = process_bug_descriptions(csv_filename, num_descriptions=100)

    # Display results
    print("\n📊 DataFrame Info:")
    print(result_df.info())

    print("\n🔍 First few rows:")
    print(result_df.head())

    # Save to new CSV
    output_filename = 'parsed_bug_descriptions.csv'
    result_df.to_csv(output_filename, index=False)
    print(f"\n💾 Saved to {output_filename}")

    print("Parser ready! Use process_bug_descriptions() function to process your CSV.")
    print("Example: result_df = process_bug_descriptions('your_file.csv', 10)")
```

Create a function to remove special characters, convert text to lowercase, and remove stop words from a given string.

```python
import re
from nltk.corpus import stopwords

def clean_text(text):
    """
    Removes special characters, converts text to lowercase, and removes stop words.

    Args:
        text: The input string.

    Returns:
        The processed string.
    """
    if not isinstance(text, str):
        return ""

    # Convert to lowercase
    text = text.lower()

    # Remove special characters
    text = re.sub(r'[^a-z\s]', '', text)

    # Split into words
    words = text.split()

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]

    # Join words back into a string
    return " ".join(words)
```
